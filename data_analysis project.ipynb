#### 2. High-Profit Order Check
Write a Python function that determines whether an order is classified as "High Profit". An order is considered "High
Profit" if the profit is greater than $500.
The function should take an Order ID as input and return True if it meets the
criteria, or False otherwise.

```python
df[['Order ID','Profit']].head(50)
```

Defined a function as - high_profit_check(orderID,df) which will return True/False to classify an order as high profit.<br>
By taking an order id as input, the function loop through each row and 
check the matching order id with the Order ID column.<br>
If matches then go to next step and check profit condition,
if the profit is greater than $500-> then return True otherwise False. If no matching order found, it prints order id not found.

```python
def high_profit_check(orderID,df):
    for i in range(len(df)):
        if df.loc[i,'Order ID']==orderID:
           result=df.loc[i,'Profit']>500
           return f"High Profit -  {result}" 
    print("Order it not found")

orderID_input=int(input("Enter Order ID"))
output=high_profit_check(orderID_input,df)

print(output)
```

#### 3. Average Revenue for a Product Category
Write a Python function that takes a Product Category as input and returns the average revenue generated from
that category.

```python
df[['Product Category','Revenue']].head(50)
```

Here, gen_avg_revenue() function will perform the task of generating average revenue for each category by taking a product category as input.<br>
Filter matching product category, If category matched then calculate the average of revenue.<br>
If input category does not match - match_category will be empty and return the message


```python
def gen_avg_revenue(pdCategory,df):
    match_category=df[df['Product Category']==pdCategory]

    if  not match_category.empty:
        avg_revenue= match_category['Revenue'].mean()
        return f"Average Revenue of {pdCategory}  {avg_revenue:.2f}" 
    
    else:
     return "Category not found"

   
category_input=input("Enter Category of a Product: ")
output=gen_avg_revenue(category_input,df)

print(output)
```

#### 5. Remove Outliers
Detect and remove rows where profits (e.g., ComputerProfit , AirConditionerProfit ) are unusually high
compared to other rows.

Find the outliers using IQR method

```python
outlier_columns=['Profit','AirConditionerProfit']

df1=data[outlier_columns]
Q1=df1.quantile(0.25)
Q3=df1.quantile(0.75)

IQR=Q3-Q1

L_bound=Q1-1.5*IQR
U_bound=Q3+1.5*IQR


for i in outlier_columns:
  L=L_bound[i]
  U=U_bound[i]


  outliers=data[(data[i]<L) | (data[i]>U)]

  print(f"Outliers in {i} :")
  print(f"\nNumber of outliers in {i}: {len(outliers)}")
  print(outliers[i],"\n")
```

```python
threshold=500

for i in outlier_columns:
  data.loc[data[i]>threshold,i]=np.nan
  data[i]=data[i].fillna(data[i].median())
```

```python
data
```

**Explanation :** Outlier Detection and Replacing with Median
 
To find the outliers in ``Profit`` and ``AirConditionerProfit`` column, the **IQR method** is used here.<br>
Then computed upper and lower bounds, which helped to find the outliers. Any data point below **lower bound** or above **upper bound** is considered an outlier. From this, we get 75 outliers in ``Profit`` column and 50 outliers in ``AirconditionerProfit`` column.<br>

After outlier values ​​are identified, a threshold is set to 500 to isolate extreme values ​​that are likely to be invalid.<br>
To treat these outliers, values ​​greater than the threshold are replaced with NaN. These missing values ​​are then handled by calculating the median of their respective columns, to ensure the data integrity.

#### 1. Independent Samples T-Test: Sales Target Achievement

**Problem Statement:**
AK Enterprise's management wants to understand if achieving sales targets has a significant
impact on customer ratings. Using the dataset, conduct an independent samples t-test to determine if there is a
statistically significant difference in mean customer ratings between sales representatives who achieved their
targets and those who did not.<br>

**Requirements:**<br>
Set up appropriate null and alternative hypotheses.<br>

Use α = 0.05 as the significance level.<br>

Calculate the t-statistic and p-value.<br>

Interpret the results in the context of the business.<br>

Include a visualization comparing the distribution of customer ratings for both groups.

<br>

To perform the **Two-Sample t-test**,  
1. First define the hypotheses-

assuming that,&emsp;μ1 = Mean customer ratings for sales representatives who achieved target<br>
&nbsp;&nbsp;&emsp;&emsp;&emsp;&emsp;&emsp;&nbsp;&nbsp;&emsp;&nbsp;μ2 = Mean customer ratings for sales representatives who did not achieve target<br>

• Null Hypotheses (H₀):  μ1 = μ2 (there is no significant difference in mean customer rating between this two groups)<br>
• Alternative Hypothesis (H₁):  μ1 ≠ μ2 (there is a significant difference in mean customer rating between this two groups)


2. Set the significance level α = 0.05 representing the maximum risk willing to accept of
incorrectly rejecting the null hypothesis.

3. Perform the test

```python
from scipy.stats import ttest_ind
```

```python
reshaped_data
```

```python
achieve_target=reshaped_data[reshaped_data['SalesTargetAchieved']=='Yes']['CustomerRating']
target_not_achieved=reshaped_data[reshaped_data['SalesTargetAchieved']=='No']['CustomerRating']

#perform two sample t-test with equal variances
t_statis,p_value=ttest_ind(achieve_target, target_not_achieved)

# print results
print(f"t-statistic: {t_statis:.4f}")
print(f"P-value: {p_value:.4f}")
```

```python
#Interpretation
alpha=0.05
if p_value <= alpha:
    print("Reject the null hypothesis: There is a significant difference in customer ratings")
else:
    print("Fail to reject the null hypothesis: No significant difference found in customer ratings")    
```

```python
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.violinplot(data=reshaped_data, x='SalesTargetAchieved', y='CustomerRating', hue='SalesTargetAchieved', palette='Set2', legend=False)
plt.title('Customer Rating Distribution by Sales Target Achievement')
plt.xlabel('Sales Target Achieved')
plt.ylabel('Customer Rating')
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Explanation:<br>**
Conducted two sample t-test using ttest_ind() from scipy.stats to examine whether there is a significant difference in mean customer ratings between sales representatives who achieved their targets and those who did not.<br>
**achieve_target** and **target_not_achieved** are two groups being compared based on whether sales target achieved or not.
Print the calculated value of t_statistic and p_value. Then compared the p-value to a significance level of 0.05.

#### Interpretation of the Results in the context of the business
From the test, the t_statistic is -0.7386 and the corresponding two-sided p-value is 0.4602.<br>
Since the p-value of our test (0.4602) is greater than alpha = 0.05, we fail to reject the null hypothesis of the test. We do not have sufficient evidence to say that there is a significant difference in average customer rating between sales representatives to achieve target.
So in the context of business we can that hitting sales targets does not appear to influence how customers rate their experience with the sales representative. Customer satisfaction may be influenced by other factors such as product quality, sales support etc.

To perform **One-sample t-test**,

1. First set the appropriate null and alternative hypotheses.<br>

 • Null Hypotheses (H₀): μ = $26   (the average profit in North region is equal to $26)<br>

 • Alternative Hypothesis (H₁):  μ < $26  (the average profit in North region is less than $26)

```python
from scipy.stats import ttest_1samp
```

2.  Use α = 0.05 as the significance level.

3. Calculate the t-statistic and p-value.

```python
N_profit=reshaped_data[reshaped_data['Region']=='North']['Profit']

t_statistic, p_value = ttest_1samp(N_profit, popmean=26)

print(f"t-statistic: {t_statistic:.4f}")
print(f"p-value: {p_value:.4f}")
```

4. **Interpretation of the result**

```python
#Interpretation
alpha = 0.05

if N_profit.mean()<26 and p_value <= alpha:
    print("Reject the null hypothesis: The average profit is significantly less than $26 per sales in North region")
else:
    print("Fail to reject the null hypothesis: No significant difference found that the average profit is less than $26 per sales in North region")    
```



From one-sample t-test we get t-statistic as -0.5768 and p-value of 0.5642<br>
Since the p value is greater than our significance level α = 0.05, so fail to reject the null hypothesis. There is no sufficient evidence to suggest that the average profit per sale in the North region is less than $26. As I found no statistical evidence to contradict it,so management's expectation of $26 profit per sale in the North region is supported.

 • Null Hypotheses (H₀): The actual distribution of high-profit sales across products follows the expected distribution. (each product accounts for 25% of high-profit sales).

 • Alternative Hypothesis (H₁): The actual distribution does not follow the expected distribution.
